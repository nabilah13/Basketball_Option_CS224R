```python
import matplotlib.pyplot as plt

rewards = ['LSPI Reward', 'DQN Reward', 'Max Reward', 'Positive Reward', 'Q2 Reward', 'Q4 Reward']
values = [2.984886156, 3.007037854, 3.987953356, 2.315396336, 1.866092776, 2.560726251]
bars = plt.bar(rewards, values, color='lightblue')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')
plt.title('2019-20 Reward Comparison')
plt.xlabel('Policy')
plt.ylabel('Reward')
plt.xticks(rotation=45)
plt.ylim(0, 4.5)
plt.tight_layout() # Adjust layout to make sure everything fits
plt.savefig('charts/reward_comparison_19_20.png')
```
